```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name"])
```

```python
District_Summary_dicts = ({"Total Schools":[school_data["School ID"].count()], 
                           "Total Students":[len(student_data["Student ID"])],
                           "Total Budget":[(school_data["budget"].sum())],
                           "Average Math Score":[student_data["math_score"].mean()],
                           "Average Reading Score":[student_data["reading_score"].mean()],
                           "% Passing Math":[((student_data["math_score"] >= 70).value_counts()[1])*100/(len(student_data["Student ID"]))],
                           "% Passing Reading":[((student_data["reading_score"] >= 70).value_counts()[1])*100/(len(student_data["Student ID"]))],
                           "% Overall Passing Rate":[(student_data["math_score"].mean()+student_data["reading_score"].mean())/2]})
District_Summary_df = pd.DataFrame(District_Summary_dicts)
District_Summary_df["Total Budget"] = District_Summary_df["Total Budget"].map("${:,.2f}".format)
District_Summary_df
```

```python
school_merge_pass=pd.merge(student_data[student_data["math_score"] >= 70], school_data, on=["school_name"])
school_merge_pass_r=pd.merge(student_data[student_data["reading_score"] >= 70], school_data, on=["school_name"])
```

```python
school_group_pass=school_merge_pass.groupby("school_name")
school_group_pass_r=school_merge_pass_r.groupby("school_name")
```

```python
school_group=school_data.groupby("school_name")
school_complete_group = school_data_complete.groupby(["school_name"])
Top_Schools1= pd.DataFrame(school_complete_group["type"].value_counts())
Top_Schools1 = Top_Schools1.rename(columns={"type": "Total Students"})
Top_Schools1 = Top_Schools1.reset_index()
Top_Schools2 = school_complete_group[["reading_score", "math_score"]].mean()
```

```python
Top_Schools2["Total School Budget"]=school_group["budget"].sum()
Top_Schools2["Per Student Budget"]= school_group["budget"].sum()/school_complete_group["student_name"].count()
```

```python
Top_Schools2["% Passing Math"] =100* school_group_pass["student_name"].count()/school_complete_group["student_name"].count()
Top_Schools2["% Passing Reading"] =100* school_group_pass_r["student_name"].count()/school_complete_group["student_name"].count()
Top_Schools2["% Overall Passing Rate"]=(Top_Schools2["% Passing Math"]+Top_Schools2["% Passing Reading"])/2
```

```python
Top_Schools = Top_Schools1.merge(Top_Schools2, on="school_name")
Top_Schools = Top_Schools.reset_index(drop=True)
Top_Schools = Top_Schools.rename(columns={"reading_score":"Average Reading Score",
                                          "math_score":"Average Math Score",
                                          "type":"School Type"})
```
